**1. Univariate Analysis**

Topics
1. Numerical : Mean,Median,Mode,variance,std
2. Skewness and kurtosis
3. Histogram and Boxplots
4. Categorical: Frequency Table, Countplots 

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)
pd.set_option("display.max_columns",50)
sns.set_style("whitegrid")
```

```python
df = pd.read_csv("titanic_dataset.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
df.columns
```

```python
num_cols = ["age","fare","sibsp","parch"]
cat_cols = ["survived","pclass","sex","embarked","class","who","adult_male","deck","embark_town","alone"]

print("Numerical Columns:",num_cols)
print("Category Columns:", cat_cols)
```

**2. Numerical Features Analysis**

```python
df[["age","fare"]].head()
```

**2.1 Summary Statistics**

```python
print("Summary Statistics:")
df[num_cols].describe()
```

**2.2 Mean, Median and Mode**

```python
print("Mean:")
print(df[num_cols].mean())

print("-"*80)

print("Median:")
print(df[num_cols].median())

```

```python
print("Mode:")
print(df[num_cols].mode().iloc[0])
```

**2.3 Variance and Standard Deviation**

```python
print("Variance:")
print(df[num_cols].var())

print("-"*85)

print("Standard Deviation:")
print(df[num_cols].std())
```

**2.4 Skewness and Kurtosis**

```python
print("Skewness:")
print(df[num_cols].skew())

print("-"*85)

print("Kurtosis:")
print(df[num_cols].kurtosis())
```

**2.5 Histograms**

```python
plt.figure(figsize=(5,3))
sns.histplot(df["age"],kde = True)
plt.show()
```

```python
for col in num_cols:
    plt.figure(figsize=(5,3))
    sns.histplot(df[col],kde = True)
    plt.title(f"Histogram: {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

**2.6 Box Plots**

```python
for col in num_cols:
    plt.figure(figsize=(5,3))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot: {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

**3. Categorical Features**

```python
df[cat_cols].head()
```

**4.1 Frequency Table**

```python
print(df["pclass"].value_counts())
print("-"*120)
print(df["pclass"].value_counts(normalize=True))
```

```python
for col in cat_cols:
    print(f"Frequency table for {col}")
    print(df[col].value_counts())
    print(f"Proportion for {col}")
    print(df[col].value_counts(normalize=True))
    print("-"*120)
```

**4.2 Countplot**

```python
for col in cat_cols:
    plt.figure(figsize=(5,3))
    sns.countplot(x=df[col])
    plt.title(f"Countplot: {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```
