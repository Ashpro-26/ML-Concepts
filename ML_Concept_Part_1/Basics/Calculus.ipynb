```python
import numpy as np
import matplotlib.pyplot as plt
np.set_printoptions(suppress=True)
```

**1. A Simple Line y = 2x + 1**

```python
def line_func(x):
    return 2 * x + 1
```

```python
x = np.linspace(-3 , 3 ,100)
y = line_func(x)
```

```python
print(x[:5])
print(y[:5])
```

```python
plt.figure(figsize=(6,4))
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(alpha = 0.3)
plt.show()


x1, x2 = -2, 1
y1, y2 = line_func(x1), line_func(x2)

rise = y2 - y1
run = x2 - x1

slope = rise / run

print("-"*50)
print(f"First Point:{x1},{y1}")
print(f"Second Point:{x2},{y2}")
print("Rise:",rise)
print("Run:",run)
print("Slope:",slope)
print("-"*50)
```

**2. Curve: y= x^2**

```python
def curve(x):
    return x**2

x_vals = np.linspace(-3,3,200)
y_vals = curve(x_vals)

plt.figure(figsize=(6,4))
plt.plot(x_vals,y_vals)
plt.title("Curve: y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(alpha = 0.3)
plt.show()
```

**Estimating Slope with a small change in x**

```python
def slope_estimate(func,x1,h=0.00001):
    x2 = x1 + h
    y1 = curve(x1)
    y2 = func(x2)
    dy = y2 - y1
    dx = h
    slope = dy / dx
    return slope
```

```python
points = [-2,0,2]
print("Slope estimate for y = x^2 using small change in x")

for p in points:
    s=slope_estimate(curve,p)
    print(f"x = {p} Estimated slope = {s:.2f}")
```

**Weight Updation:**

w = w - lr*(dL/dw)

**If dL/dw is positive increasing w increses loss so we decrease w and vice-versa**

---

```python
# Sample Dataset

x = np.array([1, 2, 3])
y_true = np.array([2, 4, 6])

w = 1.0  #Weights
lr = 0.01  #Learning Rate

def loss(w):
    y_pred = w * x
    return np.mean((y_pred - y_true)**2)

def dloss(w):
    y_pred = w * x
    return np.mean(2 * x * (y_pred - y_true))

print("Start w:", w, "Loss:", loss(w))

for step in range(30):
    slope = dloss(w)
    w = w - lr * slope
    print("Step", step+1, "w:", round(w,3), "loss:", round(loss(w), 3))

w = 1.0

ws = []
ls = []

for step in range(30):
    ws.append(w)
    ls.append(loss(w))
    slope = dloss(w)
    w = w - lr * slope
    print("Step", step+1, "w:", round(w,3), "loss:", round(loss(w), 3))

#Prepare Plot
ws_plot = np.linspace(0, 4, 200)
ls_plot = [loss(wp) for wp in ws_plot]

plt.figure(figsize=(6,4))
plt.plot(ws_plot, ls_plot, label = "Loss curve")
plt.scatter(ws, ls, color="red", label="Steps of Gradient Descent")
plt.plot(ws, ls, color = "red")
plt.legend()
plt.xlabel("w")
plt.ylabel("Loss")
plt.title("Gradient Descent on Loss")
plt.grid(alpha = 0.3)
```
