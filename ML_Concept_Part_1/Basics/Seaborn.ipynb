```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Load inBuilt Seaborn Dataset**

```python
tips = sns.load_dataset("tips")
```

```python
tips.shape
tips.head()
```

```python
#Set Global Style
sns.set_style("darkgrid")
sns.set_palette("viridis")
```

**Histogram - Distribution of Toatal Bill**

```python
plt.figure(figsize=(6,4))
sns.histplot(tips["total_bill"],kde=True)
plt.title("Distrbution of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

**Count Plot - Number of Records per Day**

```python
plt.figure(figsize=(6,4))
sns.countplot(x="day", data=tips, hue="sex", palette="Set2")
plt.title("Number of Records per day")
plt.show()
```

**Box Plot - Detect Outliers**

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="day",y="total_bill",data=tips,palette="pastel",hue="day")
plt.title("Total Bill by Day")
plt.show()
```

**Scatter Plot - Tips based on Gender**

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="total_bill",y="tip",data=tips,hue="sex",palette="cool")
plt.title("Bill v/s Tip")
plt.show()
```

**Pair Plot - Pair-wise Feature Relationship**

```python
sns.pairplot(tips,hue="sex",diag_kind="kde")
plt.suptitle("Pair-wise Relationship")
plt.show()
```

**HeatMap - Feature Correlation**

```python
correlation_matrix = tips.corr(numeric_only=True)
plt.figure(figsize=(6,4))
sns.heatmap(correlation_matrix,annot=True,cmap="coolwarm",linewidths=0.5)
plt.title("Feature Correlation HeatMap")
plt.show()
```
