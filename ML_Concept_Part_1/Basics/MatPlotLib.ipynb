**Install and Import MatPlotLib**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Simple Line Plot**

```python
x = np.linspace(0,10,50)
y = np.sin(x)

print(x[:5])

plt.plot(x,y,color="red",linestyle="--",marker="o",label="sin(x)")
plt.title("Simple Line Plot")
plt.xlabel("Angles")
plt.legend()
plt.ylabel("Corresponding Sine Values")
plt.grid()
plt.show()
```

**Scatter Plot**

```python
np.random.seed(42)

x = np.random.randn(50)
y = np.random.randn(50)

plt.scatter(x,y,color = "purple",alpha = 0.67)
```

**Histogram Plot**

```python
data = np.random.randn(1000)

plt.hist(data,bins=30,color="skyblue",edgecolor="black")
plt.title("Histogram Plot")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

**Pie Charts**

```python
labels = ["Python","Java","C++","R"]
sizes = [45, 30, 15, 10]
plt.pie(sizes,labels=labels, autopct="%1.1f%%");
plt.show()
```

**Bar Chart**

```python
students=["Alice","Bob","Jake","Anthony"]
marks=[68,84,91,72]
plt.bar(students,marks,color="orange")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```
