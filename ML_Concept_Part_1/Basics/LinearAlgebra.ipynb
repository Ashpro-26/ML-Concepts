**Scaler, Vector, Matrix and Tensor**

```python
scaler = 5.0;
print(scaler)

print("-"*40)

vector = np.array([3,7,2])
print("Vector:")

print(vector)

print("Shape of vector",vector.shape)

print("-"*40)

matrix = np.array([
    [1,2,3],
    [4,5,6]
])

print(matrix)
print("Shape of Matrix:",matrix.shape)
```

```python
v1 = np.array([1,3,5])
v2 = np.array([4,2,6])

print("v1:",v1)
print("v2:",v2)
diff = v2 - v1
dot = np.dot(v1,v2)
print("Dot Product of the vectors v1 and v2:",dot)
print("Difference of the vectors v1 and v2:",diff)
```

**Eucledian Distance**

```python
diff = v2 - v1
print("Difference of the vectors v1 and v2:",diff)

difference = np.linalg.norm(diff)
print(difference)
```

**Matrix Multiplication and Linear Model**

```python
x= np.array(
    [[1.0,30.0],
     [2.0,50.0],
     [3.0,70.0]]
)

print("x:\n",x)
print(x.shape)
```

```python
w = np.array([
    [0.5],
     [0.1]
])

print("w:\n",w)
print(w.shape)
```

```python
xw = x@w

print(xw)
```

**Connecting to a Sample Dataset**

```python
data = np.array([
    [85,26.6,31,0],
    [183,23.3,32,1],
    [89,28.1,21,0],
    [137,43.1,33,1]
])

print("\n Full Data Matrix:\n",data)
print("Data Shape:",data.shape)
```

```python
x_features = data[:, :-1]
y_target = data[:, -1]

print("\nFeature matrix X:\n",x_features)
print("\nX Shape:\n",x_features.shape)

print("\nTarget vector Y:\n",y_target)
print("\nY Shape:\n",y_target.shape)
```
